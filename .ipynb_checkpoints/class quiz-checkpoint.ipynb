```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
warnings.simplefilter("ignore")
```

```python
df = pd.read_csv('TMeURO.csv')
```

```python
df.drop(['Latitude', 'Longtitude'], axis=1, inplace = True)
df
```

```python
df.rename(columns={'City Code': 'City'}, inplace=True)
df
```

```python
df.rename(columns={'Order Priority Code': 'Order Priority'}, inplace=True)
df
```

```python
df[df.duplicated()]
```

```python
df = df.drop_duplicates()
df
```

```python
df[df.duplicated()]
```

```python
df.isnull().sum()
```

```python
import numpy as np
```

```python
city = df['City'].median()
city
```

```python
order_priority = df['Order Priority'].median()
```

```python
order_priority
```

```python
df['City'].replace(to_replace = np.nan, value = city, inplace = True)
```

```python
df['Order Priority'].replace(to_replace = np.nan, value = order_priority, inplace = True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
city_code_mapping = {
    1: 'Karachi',
    2: 'Lahore',
    3: 'Islamabad',
    4: 'Faisalabad',
    5: 'Quetta',
    6: 'Peshawar'
}

df['City'] = df['City'].map(city_code_mapping)
df
```

```python
sorted_products = df.groupby('product')['total'].sum().sort_values(ascending = False).head(5).reset_index()
sorted_products
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(x='product', y='total', data=sorted_products, palette='viridis')
plt.title('Top 5 Best Selling Products')
plt.xlabel('Total Sales')
plt.ylabel('Product')
plt.show()
```

**Based on the analysis, the chart reveals the top five best-selling products in terms of total sales. It shows that headphones lead the sales, followed closely by watches and mobiles, while laptops and shoes have lower sales figures in comparison.**


```python
payment_distribution = df['payment_method'].value_counts()

plt.figure(figsize=(6, 6))
payment_distribution.plot.pie(autopct='%1.1f%%', startangle=90, colors=['#FF9999', '#66B2FF', '#99FF99', '#FFCC99', '#FFD700'])
plt.title('Payment Method Distribution')
plt.ylabel('')
plt.show()
```

**This pie chart shows the payment method distribution based on the total sales. JazzCash has the highest share at 22.4%, followed by Credit Card and COD with shares of 21.4% and 19.4%, respectively. Debit Card and Easypaisa have smaller portions, at 15.8% and 21.0%, respectively, illustrating how each payment method contributes to the overall sales volume.**

```python
df['date'] = pd.to_datetime(df['date'], format='%d/%m/%Y')
df['Month'] = df['date'].dt.month_name()
monthly_sales = df.groupby('Month')['total'].sum().reindex(['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])

plt.figure(figsize=(8, 4))
monthly_sales.plot(kind='line', marker='o', color='b')
plt.title('Monthwise Sales')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

**This line chart shows monthly sales, with total sales plotted against each month. The chart highlights a significant peak in March, followed by a sharp decline in May, and then steady growth towards November, indicating fluctuating sales trends over the months.**

```python
city_sales = df.groupby('City')['total'].sum()
city_sales = city_sales.sort_values(ascending=False)

plt.figure(figsize=(8, 4))
sns.barplot(x=city_sales.index, y=city_sales, palette='viridis')
plt.title('City Wise Sales')
plt.xlabel('Total Sales')
plt.ylabel('City')
plt.show()
```

**This bar chart represents the city-wise sales. Lahore has the highest total sales, followed by Quetta and Islamabad, while cities like Karachi and Peshawar have the lowest sales figures, providing a clear comparison of sales performance across different cities.**

```python
df['Month'] = pd.to_datetime(df['date'], format='%d/%m/%Y').dt.month_name()

monthly_category_sales = df.groupby(['Month', 'category'])['total'].sum().unstack()

monthly_category_sales = monthly_category_sales.loc[monthly_category_sales.sum(axis=1).sort_values(ascending=False).index]

monthly_category_sales.plot(kind='bar', figsize=(10, 5), width=0.8)
plt.title('Monthly Sales by Category')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

**This bar chart displays the monthly sales by category, where the total sales of Accessories, Electronics, and Fashion are compared across the months. The chart shows that Accessories and Electronics sales peak in March and April, while Fashion shows more consistent sales throughout the year, with a noticeable dip in July and August.**

## CONCLUSION

**The analysis shows that headphones are the top-selling product, followed by watches and mobiles, indicating strong interest in tech and fashion items. Sales were highest in March and November, with Lahore and Quetta driving the most sales. JazzCash is the most popular payment method, followed by Credit Cards and COD. These insights can help the business focus on popular products, target key cities for sales, and offer preferred payment methods to improve customer experience and boost sales.**
