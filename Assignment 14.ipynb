#### Check Duplicate Records

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("cars - cars.csv")
df
```

```python
df.duplicated().sum()
```

#### Drop Duplicate Records

```python
df = df.drop_duplicates().reset_index(drop=True)
df
```

#### Check missing or null values

```python
df.isnull().sum()
```

#### Drop missing or null values

```python
df = df.dropna().reset_index(drop=True)
df
```

#### Show top 10 average MSRP (Price) per car brand (rounded figure)?

```python
df.groupby('Make')['MSRP'].mean().round().sort_values(ascending=False).head(10)
```

#### Show the average MSRP (price) per years (rounded figure)?

```python
df.groupby('Year')['MSRP'].mean().round()
```

#### Show average Engine HP (power) rounded figure and MPG (efficiency) rounded figure per year?

Hint: Average of (city mpg + highway MPG)

```python
df['Average MPG'] = ((df['city mpg'] + df['highway MPG']) / 2).round()
df.groupby('Year')[['Engine HP', 'Average MPG']].mean().round()
```

#### What percentage of cars are Compact, Midsize, or Large?

```python
sizes = df['Vehicle Size'].dropna().str.strip()
sizes.isin(['Compact', 'Midsize', 'Large']).mean() * 100
```

#### Which vehicle styles are the most sold (most common) in the dataset?

```python
df['Vehicle Style'].value_counts().head(10)
```

#### Which car brands have the most vehicles priced at $40,000 or more and with a popularity score of at least 1000?

```python
df[(df['MSRP'] >= 40000) & (df['Popularity'] >= 1000)]['Make'].value_counts().head(10)
```
